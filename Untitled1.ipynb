```python

print("== CLI 게시판 실행 ==")
i = 1
j = 1
while True:
    cmd = input("명령어 ) ")

    if cmd == "exit":
      break

    elif cmd == "article write":
      input("제목 : ")
      input("내용 : ")
      print("{}번 글이 생성되었습니다".format(i))
      i += 1

    elif cmd == "article list": #딕셔너리나 리스트로 저장, 저장한 다음에 꺼내써야함
      print("===========================")
      print("번호   /   제목   /   내용")
      while j <= i:
        print("{j}   /   {}   /   {}".format(j, ,))
        j += 1
      print("===========================")

    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")
```
