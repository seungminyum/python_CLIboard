```python

print("== CLI 게시판 실행 ==")
i = 1
articles =[] #게시글 저장소

while True:
    cmd = input("명령어 ) ").strip() #공백을 지워주는 요소


    if cmd == "exit":
      break

    elif cmd == "article write":
      id = i
      title = input("제목 : ")
      body = input("내용 : ")
      article = {"id" : id, "title" : title, "body" : body} #조립 != 저장
      #print(article)
      articles.append(article) #저장
      print("{}번 글이 생성되었습니다".format(i))
      i += 1

    elif cmd == "article list": #딕셔너리나 리스트로 저장, 저장한 다음에 꺼내써야함
      if not articles:
        print("게시글이 없습니다")
      else:
        print("===========================")
        print("   번호   /   제목   /   내용")
        for a in reversed(articles):
          print("  {}   /    {}    /   {}".format(a["id"], a["title"], a["body"]))
      print("===========================")

    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")
```
